# Exploring the Data

## I am taking two criteria to explore the data 
1. Distribution
2. Profit

## Plotting graphs to check the Distribution of the Products 
### 1. Based on Category

```python
plt.figure(figsize=(8,5))
plt.title('DISTRIBUTION OF CATEGORY')
sns.countplot(x='Category', data = data)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.show()
```

### 2. Based on Regions

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot('Region',data=data)
plt.title('DISTRIBUTION OF REGIONS')
plt.xlabel('REGION',fontsize=18)
plt.ylabel('COUNT',fontsize=18)
plt.grid(alpha=0.1)
```

### 3. Distribution of Regions based on the Category

```python
plt.figure(figsize=(20,6))
plt.subplot(1,2,2)
sns.countplot('Region',hue='Category',data=data)
plt.title('DISTRIBUTION OF REGIONS BASED ON CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.1)
plt.show()
```

### 4. Distribution based on States

```python
plt.figure(figsize=(20,5))
sns.countplot(data['State'])
plt.xticks(fontsize=16,rotation=90)
plt.ylabel('COUNTS',fontsize=20)
plt.xlabel('STATES',fontsize=20)
plt.title('DISTRIBUTION OF STATES',fontsize=25);
plt.grid(alpha=0.3)
plt.show()
```

## Now Let' s Analyse the data based on Profit 
### 1. Segment Vs Profit 

```python
sns.catplot(x="Segment",y="Profit",data=data,kind='bar',ci=None)
```

### 2. Shipmode Vs Profit

```python
sns.catplot(x="Ship Mode",y="Profit",data=data,kind='bar',ci=None)
```

### 3. State Vs Profit

```python
order = data.groupby(["State"])["Profit"].mean().sort_values().index
sns.catplot(y="State",x="Profit",data=data,kind='bar',ci=None,height=10,order=order)
```

### 4. Now let's see in which category, we have the maximum Profit (Category vs Profit)

```python
sns.catplot(x="Category",y="Profit",data=data,kind="bar",ci=None)
```

### 5. Profit vs Category (In different regions)

```python
sns.catplot(x="Category",y="Profit",data=data,kind='bar',ci=None,col="Region")
```
